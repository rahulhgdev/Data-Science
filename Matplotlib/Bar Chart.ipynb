```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
company = ['Rel','IO ','O&N','SBI','BPC','TM','RE','TCS','ICICI','L&T']
revenue = [990,880,770,660,550,440,330,220,110,50]
profit = [510,490,420,370,320,310,200,150,70,30]
len(profit)
```

```python
xpos = np.arange(len(company))
```

```python
xpos
```

```python
plt.xticks(xpos,company)   # xticks will replace ypos's integer with company name
plt.ylabel('Revenue(billion $)')
plt.title('Top Indian Companies by Revenue Forbes')

plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label='Profie')
plt.legend()
```

## For side-by-side comparison

```python
plt.xticks(xpos,company)           # xticks will replace xpos's integer with company name
plt.ylabel('Revenue(Nill $)')
plt.title('Top Indian Companies by Revenue Forbes')

plt.bar(xpos-0.2,revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2,profit, width=0.4, label='Profie')
plt.legend()

# to save

plt.savefig('Output/bar_output.png', dpi=200)
```
